# Grafik Training

```python
import os
result_path = '/content/drive/MyDrive/Fitria/Result2/vox2vox/fix1/3batch'
os.listdir(result_path)
```

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
history = pickle.load(open(result_path + '/history_last.pkl', 'rb'))
print('Last epoch:', history['last_epoch'])
train_history = pd.DataFrame.from_dict(history['train'][:100])
train_history.index += 1
valid_history = pd.DataFrame.from_dict(history['valid'][:100])
valid_history.index += 1
```

```python
train_history.v2v.plot(title='Generator Loss')
valid_history.v2v.plot(title='Generator Loss')
plt.legend(['training', 'validation'])
plt.show()

train_history.disc.plot(title='Discriminator Loss')
valid_history.disc.plot(title='Discriminator Loss')
plt.legend(['training', 'validation'])
plt.show()

train_history.dice.plot(title='Dice Loss')
valid_history.dice.plot(title='Dice Loss')
plt.legend(['training', 'validation'])
plt.show()

train_history.iou.plot(title='IOU Loss')
valid_history.iou.plot(title='IOU Loss')
plt.legend(['training', 'validation'])
```
